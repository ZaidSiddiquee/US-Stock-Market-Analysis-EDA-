```python
tech_companies = ["Apple_Price", "Apple_Vol.", "Tesla_Price", "Tesla_Vol.", "Microsoft_Price", "Microsoft_Vol.",
                  "Google_Price", "Google_Vol.", "Nvidia_Price", "Nvidia_Vol.", "Berkshire_Price", "Berkshire_Vol.",
                  "Netflix_Price", "Netflix_Vol.", "Amazon_Price", "Amazon_Vol.", "Meta_Price", "Meta_Vol."]

tech_data = df[["Date"]+ tech_companies]
tech_data.set_index("Date", inplace = True)
tech_data.head()
```

FAANG Companies Price trend over time:

```python
plt.figure(figsize = (12,6))
for col in ['Meta_Price', 'Apple_Price', 'Amazon_Price', 'Netflix_Price', 'Google_Price']:
  plt.plot(tech_data.index,tech_data[col], label = col)

plt.title("FAANG companies price over time")
plt.xlabel("Date")
plt.ylabel("Price")
plt.legend()
plt.grid()
plt.show()
```

```python
selected_companies = ['Apple_Price', 'Tesla_Price', 'Microsoft_Price', 'Google_Price', 'Nvidia_Price',
                      'Meta_Price', 'Netflix_Price', 'Amazon_Price', 'Berkshire_Price']

selected_volumns = [col.replace("Price", "Vol.") for col in selected_companies]
tech_data = df[["Date"]+ selected_companies + selected_volumns]
tech_data.head()
```

```python
tech_data[selected_companies].describe()
```

.

.

.

Calculating average prices and volumes

```python
Average_com_Price = tech_data[selected_companies].mean()
Average_com_Volumes = tech_data[selected_volumns].mean()

# Combined bar plot for average prices and volumes
fig, axl = plt.subplots(figsize = (12,6))

# Bar plot for average prices
Average_com_Price.plot(kind = "bar", color = "skyblue", ax = axl, position=0, width=0.4, label='Average Price')
axl.set_ylabel("Average_Price", color = "blue")
axl.tick_params(axis = "y", labelcolor = "blue")


# Bar plot for average volumes (on secondary y-axis)
ax2 = axl.twinx()
Average_com_Volumes.plot(kind = "bar", color = "orange", ax = ax2, position = 1, width = 0.4, label = "Average_Volume")
ax2.set_ylabel("Average Volume" ,color = "orange")

plt.title("Comparision of Average Price and Volumes")
plt.xticks(rotation = 45)
fig.tight_layout()
plt.show()
```

.

.

Drawing correlation matrix:

```python
Com_price_corr = tech_data[selected_companies].corr()

plt.figure(figsize = (10,6))
sns.heatmap(Com_price_corr, annot = True, cmap = "coolwarm", fmt = ".2f")
plt.title("Correlation between company prices")
plt.show()
```

```python
tech_data.head()
```

Stock Price trends for all companies:

```python
Company_Price = ['Apple_Price', 'Tesla_Price', 'Microsoft_Price', 'Google_Price', 'Nvidia_Price',
                  'Meta_Price', 'Netflix_Price', 'Amazon_Price', 'Berkshire_Price']


# Create subplots for time-series trends
num_companies = len(Company_Price)
rows = (num_companies + 2) // 3   # Adjust number of rows for subplots
fid, axes = plt.subplots(rows, 3, figsize = (15, rows*4))

# Flatten axes for easy indexing
axes = axes.flatten()

# Plot each company's time-series data
for idx, company in enumerate(Company_Price):
    ax = axes[idx]
    ax.plot(tech_data.index, tech_data[company], label=company.split('_')[0], color='tab:blue')
    ax.set_title(f"{company.split('_')[0]} Price Trend")
    ax.set_xlabel('Date')
    ax.set_ylabel('Price')
    ax.legend()
    ax.grid()

# Hide any extra subplots (if rows * cols > num_companies)
for i in range(num_companies, len(axes)):
    axes[i].set_visible(False)

# Adjust layout
plt.tight_layout()
plt.suptitle("Time-Series Trends for Tech Companies", fontsize=16, y=1.02)
plt.show()

```
